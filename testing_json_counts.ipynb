```python
import pandas as pd
import json

# Load the existing leaderboard JSON file
with open("leaderboard_0220.json", "r") as f:
    leaderboard_data = json.load(f)

# Load the CSV file with team information.
# The CSV is expected to have columns: 'team', 'institute', and 'country'
df_team_info = pd.read_csv("Phase1_info.csv", usecols=["team", "Institution", "Country"])


df_team_info['team'] = df_team_info['team'].str.strip()

# Convert the desired columns to string type.
df_team_info["team"] = df_team_info["team"].astype(str)
df_team_info["Institution"] = df_team_info["Institution"].astype(str)
df_team_info["Country"] = df_team_info["Country"].astype(str)


# Convert the leaderboard JSON data to a DataFrame
df_leaderboard = pd.DataFrame(leaderboard_data)

# Remove any whitespace from the team names in the CSV (e.g., "Team AIMS" becomes "TeamAIMS").
df_team_info["team"] = df_team_info["team"].str.replace(r"[\s@-]+", "", regex=True)


# Merge the leaderboard data with the team info on the "team" column.
df_merged = pd.merge(df_leaderboard, df_team_info, on="team", how="left")

# (Optional) Reorder or format the DataFrame as needed.
# For example, you might want to display the institute and country below the team name in your UI.
# Here, we simply keep them as separate keys.

# Convert the merged DataFrame back to a list of dictionaries
merged_json = df_merged.to_dict(orient="records")
# Remove duplicates due to merge based on the team column.
df_merged = df_merged.drop_duplicates(subset=["team"], keep="first")

# Create a combined column for team, Institution, and Country.
df_merged["Team Info"] = df_merged["team"] + " (" + df_merged["Institution"] + ", " + df_merged["Country"] + ")"

# (Optional) If you prefer to drop the separate columns after combining:
df_merged = df_merged.drop(columns=["Institution", "Country"])



# Save the new JSON to a file
with open("leaderboard_with_info.json", "w") as f:
    json.dump(merged_json, f, indent=4)

print("New JSON file 'leaderboard_with_info.json' saved with institute and country info appended.")

df_team_info
```
